# Check Milvus Vector Database

This notebook verifies that the Milvus Lite `.db` file was created and contains chunks.

```python
from pathlib import Path

# Check if data/milvus.db exists
db_path = Path("../data/milvus.db")
print(f"DB path: {db_path.absolute()}")
print(f"DB exists: {db_path.exists()}")

if db_path.exists():
    print(f"DB file size: {db_path.stat().st_size} bytes")
```

```python
# Connect to Milvus Lite
from pymilvus import MilvusClient

client = MilvusClient("../data/milvus.db")
print("Connected to Milvus Lite")
```

```python
# List all collections
collections = client.list_collections()
print(f"Collections: {collections}")

if "chunks" in collections:
    print("✓ 'chunks' collection exists")
    # Load the collection into memory before querying
    client.load_collection("chunks")
    print("✓ Collection loaded")
```

```python
# Check collection stats
if "chunks" in collections:
    stats = client.get_collection_stats("chunks")
    print(f"Collection stats: {stats}")
```

```python
# Query sample chunks
if "chunks" in collections:
    # Get total count
    result = client.query(
        "chunks",
        filter="chunk_index >= 0",
        limit=100,
        output_fields=[
            "chunk_id",
            "source_url",
            "page_title",
            "text",
            "char_count",
            "price_gbp",
        ],
    )

    print(f"Total chunks stored: {len(result)}")
    print("\nFirst 5 chunks:")
    for i, chunk in enumerate(result[:5]):
        print(f"\n--- Chunk {i + 1} ---")
        print(f"ID: {chunk['chunk_id']}")
        print(f"URL: {chunk['source_url']}")
        print(f"Title: {chunk['page_title']}")
        print(
            f"Text: {chunk['text'][:100]}..."
            if len(chunk["text"]) > 100
            else f"Text: {chunk['text']}"
        )
        print(f"Char count: {chunk['char_count']}")
        print(f"Price: {chunk['price_gbp']}")
```

```python
# Summary
if "chunks" in collections:
    result = client.query("chunks", filter="chunk_index >= 0", limit=1)
    total_count = len(result)

    print("\n=== SUMMARY ===")
    print(f"✓ Milvus database file exists at {db_path.absolute()}")
    print("✓ 'chunks' collection created and populated")
    print(f"✓ Total chunks stored: {total_count}")
    print(
        "✓ Each chunk has: chunk_id, source_url, chunk_index, text, char_count, page_title, price_gbp, vector"
    )
else:
    print(
        "✗ No 'chunks' collection found. Make sure to run the /scrape endpoint first."
    )
```
